```python
import numpy as np
import os 
import matplotlib.pyplot as plt
from glob import glob

from utils import ResultsPlotter
```

```python
def get_files_from_replicate(channels, train_rep, test_rep):
    path_pattern = f'../DATA/AI4AB_v2_E_coli_txt_only/*Antibiotics_AvgPoolCNN_cross_val_{channels}/test_on_rep_{train_rep}/Plate_{test_rep}/'
    path = glob(path_pattern)[0]
    print(path)

    feat_vecs = np.loadtxt(os.path.join(path, 'feat_vecs.txt'))
    labels = np.loadtxt(os.path.join(path, 'labels.txt'))
    preds = np.loadtxt(os.path.join(path, 'preds.txt'))
    test_outputs = np.loadtxt(os.path.join(path, 'test_outputs.txt'))
    return feat_vecs, labels, preds, test_outputs
```

```python
feat_vecs = []
labels = [] 
preds = []
test_outputs = []
plate_id = []
channel_id = []

for ch_id, ch in enumerate(['BF', 'Hoechst_FM4_BF']):
    for p_id, rep in enumerate([1, 2, 3, 4]):
        print(ch, rep)
        feat_vecs_, labels_, preds_, test_outputs_ = get_files_from_replicate(ch, rep, rep)
    
        feat_vecs.append(feat_vecs_)
        labels.append(labels_)
        preds.append(preds_)
        test_outputs.append(test_outputs_)
        plate_id.append(np.ones_like(labels_) * p_id)
        channel_id.append(np.ones_like(labels_) * ch_id)
```

```python
feat_vecs = np.vstack(test_outputs)
labels = np.hstack(labels)
preds = np.hstack(preds)
test_outputs = np.vstack(test_outputs)
plate_id = np.hstack(plate_id)
channel_id = np.hstack(channel_id)
```

```python
plotter = ResultsPlotter(feat_vecs=feat_vecs,
                         labels=labels,
                         preds=preds,
                         plate_id=plate_id,
                         channel_id=channel_id,
                         test_outputs=test_outputs,
                         classes=classes,
                         ch_name_list=['BF', 'Hoechst_FM4_BF']
```
